```python
airfoil_self_noise_X_train = np.loadtxt(
    open(
        'airfoil_self_noise_X_train.csv',
        "rb"),
    delimiter=",",
    skiprows=1)
airfoil_self_noise_X_test = np.loadtxt(
    open(
        'airfoil_self_noise_X_test.csv',
        "rb"),
    delimiter=",",
    skiprows=1)
airfoil_self_noise_y_train = np.loadtxt(
    open(
        'airfoil_self_noise_y_train.csv',
        "rb"),
    delimiter=",",
    skiprows=1)
airfoil_self_noise_y_test = np.loadtxt(
    open(
        'airfoil_self_noise_y_test.csv',
        "rb"),
    delimiter=",",
    skiprows=1)
```

read csv

```python
airfoil_self_noise_X_test = np.mat(airfoil_self_noise_X_test)
airfoil_self_noise_X_train = np.mat(airfoil_self_noise_X_train)
m = airfoil_self_noise_X_train.shape[0]
airfoil_self_noise_y_test = np.mat(airfoil_self_noise_y_test)
airfoil_self_noise_y_train = np.mat(airfoil_self_noise_y_train)
airfoil_self_noise_y_train = np.mat(airfoil_self_noise_y_train.T)
b = np.ones(m)
airfoil_self_noise_X_train = np.c_[
    b, airfoil_self_noise_X_train]  # add the first column 1
airfoil_self_noise_X_train = np.matrix(airfoil_self_noise_X_train)
# print(airfoil_self_noise_X_train)
airfoil_self_noise_X_trainT = np.mat(airfoil_self_noise_X_train.T)  # compute T
times = np.mat(np.dot(airfoil_self_noise_X_trainT, airfoil_self_noise_X_train))
c = np.mat(np.dot(times.I,
                  airfoil_self_noise_X_trainT))  # compute coefficients
# print(c.shape)
# print(airfoil_self_noise_y_train.shape)
c = np.mat(np.dot(c, airfoil_self_noise_y_train))

print(c)
```

print c

```python
m_1, n_1 = airfoil_self_noise_X_test.shape
d_2 = 0.0
y_1 = np.ones([m_1, 1]) * c[0, 0]

y_1 = np.mat(y_1)
for i in range(0, m_1 - 1):
    for j in range(0, n_1 - 1):
        y_1[i, 0] = y_1[i, 0] + c[j + 1, 0] * airfoil_self_noise_X_test[i, j]
    d_2 = d_2 + (airfoil_self_noise_y_test[0, i] - y_1[i, 0])**2

error = (d_2 / m_1)**0.5
print(error)

# mu=average
mu_y = 0.0  # mean of true data
mu_y1 = 0.0  # mean of estimator of y

for i in range(0, m_1 - 1):
    mu_y = mu_y + airfoil_self_noise_y_test[0, i]
    mu_y1 = mu_y1 + y_1[i, 0]

mu_y = mu_y / m
mu_y1 = mu_y1 / m
```

```python
a = 0.0
b = 0.0
c = 0.0

for i in range(0, m_1 - 1):
    a = a + (airfoil_self_noise_y_test[0, i] - mu_y) * (y_1[i, 0] - mu_y1)
    b = b + (airfoil_self_noise_y_test[0, i] - mu_y)**2
    c = c + (y_1[i, 0] - mu_y1)**2

b = b**0.5
c = c**0.5
r_p = a / (b * c)
print("The Pearson correlation is %r" % r_p)
```
